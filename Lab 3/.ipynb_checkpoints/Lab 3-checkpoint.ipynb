### Выборка по датафрейму

Для наглядности сделаем выборку по следующему условию:<br>
*Длина стебля Ириса свыше 7 см (первый столбец датафрейма "5.1") и Ирис принадлежит классу "Iris-virginica"*

```python
df[(df['5.1'] > 7.0) & (df['Iris-setosa'] == 'Iris-virginica')]
```

### Корреляция по всему датафрейму

```python
df.corr()
```

### Корреляция по группам

```python
grouped_df = df.groupby('Iris-setosa').corr()
grouped_df
```
